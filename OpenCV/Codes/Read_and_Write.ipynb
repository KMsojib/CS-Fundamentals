```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("Photos/cats.jpg")

img1 = cv.imread('photos/cats.jpg', cv.IMREAD_COLOR)     # default = color
img2 = cv.imread('photos/cats.jpg', cv.IMREAD_GRAYSCALE) # 0 = grayscale
img3 = cv.imread('photos/cats.jpg', cv.IMREAD_UNCHANGED) # -1 = with alpha channel
plt.imshow(img1)
plt.axis('off')
plt.show()
```
