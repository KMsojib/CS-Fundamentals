```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread(r'D:\A - Programming & Job Preparation\CS-Fundamentals\OpenCV\Photos\cats.jpg')
if img is None:
    raise FileNotFoundError("Image not found!")

# Gray + Blur
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
blur = cv.GaussianBlur(gray, (5, 5), 0)

# Show using matplotlib
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blur, cmap='gray')
plt.title("Blurred")
plt.axis('off')

plt.show()
```

# Finding Edges & Lines

```python
# Finding Edges & Lines
img = cv.imread(r"D:\A - Programming & Job Preparation\CS-Fundamentals\OpenCV\Photos\group 1.jpg")
edges = cv.Canny(img, 50, 150)

plt.figure(figsize=(6, 6))
plt.imshow(edges, cmap='gray')
plt.title("Edges - Table Lines")
plt.axis('off')
plt.show()
```

# Masking = Drawing Black Rectanles

```python
x1,y1 = 500, 500
x2,y2 = 500, img.shape[0]

cv.rectangle(img, 
              (x1, y1), 
              (x2, y2), 
              (0, 0, 0),      
              -1)

plt.figure(figsize=(6,6))
plt.imshow(img,cmap = 'gray')
plt.title("Masking Image")
plt.axis('off')
plt.show()
```
