```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = (r'D:\A - Programming & Job Preparation\CS-Fundamentals\OpenCV\Basic Document Masker\input\invoice-1.png')
MASK_AREAS = [
    (100, 150, 500, 350),
    (550, 150, 800, 300),
    (700, 400, 900, 800)
]


# Load the images
img = cv.imread(img_path)

# Chekc image are load or not
if img is None:
    print("Error.. Image didn't load")
    exit()

# Draw Rectangle
for(x1, y1, x2, y2) in MASK_AREAS:
    cv.rectangle(img, (x1,y1),(x2,y2),(0,0,0),-1)
```

```python
# Show the ouput
plt.imshow(img,cmap='gray')
plt.title("Maksing Image")
plt.axis('off')
plt.show()
```
