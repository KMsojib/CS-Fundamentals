```python
# Repeat an array
arr = np.array([[1,2,3]])
r1 = np.repeat(arr, 3, axis=0)
print(r1)
```

```python
# fill the middle element of the matrix  = 9 and the all side element  = 1 and rest of the element will be become 0

output = np.ones((5,5))
print(output)

# make the 3,3 matrix
zero = np.zeros((3,3))
print(zero)

# initilize the midle = 9
zero[1,1] = 9
print(zero)

# Now make the full matrix
output[1:4, 1:4] = zero
print("Final Matrix\n",output)
```
