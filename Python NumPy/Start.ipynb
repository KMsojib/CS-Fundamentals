```python
import numpy as np
```

# The Basics

```python
a = np.array([1,2,3])
print(a)
```

```python
b = np.array([[1,2,3],[4,5,6]])
print(b)
```

```python
# Get the Dimension
print(a.ndim)
print(b.ndim)
```

```python
# Shape
print(a.shape)
print(b.shape)
```

```python
# Get type
print(a.dtype)
print(b.dtype)
```

```python
# Get size
a.itemsize
```

```python
# Get total Size
a.nbytes
```
